```python
import pandas as pd
```

```python
dataset=pd.read_csv('blackFriday_train.csv')
```

```python
dataset.head()
```

```python
#convert gender cate into numeric m-0,f-1:
df=dataset.copy()
```

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
```

```python
df
```

```python
df.Age.unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'0-17':2,'26-35':3,'46-50':4,'51-55':5,'36-45':6,'18-25':7})
```

```python
#creating dummies for city column
city=pd.get_dummies(df['City_Category'],drop_first=True)
city
```

```python
#concatinate city to df
df = pd.concat([df,city],axis=1)

#drop columns- City_Category
df.drop(labels='City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#value counts each 
df.Product_Category_2.value_counts()  
```

```python
#fill nan values with mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
#replacing special symbols in columns
df['Stay_In_Current_City_Years']=df.Stay_In_Current_City_Years.str.replace('+','') # replacing + with blank
```

```python
#converting datatypes into integer

df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
